### Time Series Forecasting with Machine Learning : XGBoost

https://www.kaggle.com/code/robikscube/time-series-forecasting-with-machine-learning-yt

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import xgboost as xgb

from sklearn.metrics import mean_squared_error
color_pal = sns.color_palette()
plt.style.use('fivethirtyeight')

```

```python
df = pd.read_csv('/Users/vkbharadwaj/Dropbox/Polaris/scratch/projects/AI_projects/from-skyrmions-to-smartgrids/data/AEP_hourly.csv')
df.head()
df.tail()
df = df.set_index('Datetime')
df.index = pd.to_datetime(df.index)
df = df.sort_index()
```

```python
%matplotlib inline
df.plot(style='.',
        figsize=(15, 5),
        color=color_pal[0],
        title='PJME Energy Use in MW')
plt.show()
```

```python
train = df.loc[df.index < '01-01-2015']
test = df.loc[df.index >= '01-01-2015']

fig, ax = plt.subplots(figsize=(15, 5))
train.plot(ax=ax, label='Training Set', title='Data Train/Test Split')
test.plot(ax=ax, label='Test Set')
ax.axvline('01-01-2015', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
plt.show()
```

```python

df.loc[(df.index > '01-01-2010') & (df.index < '01-8-2010')] \
    .plot(figsize=(17, 5), title='Week Of Data')
plt.show()

```

```python
def create_features(df):
    """
    Create time series features based on time series index.
    """
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    df['dayofmonth'] = df.index.day
    df['weekofyear'] = df.index.isocalendar().week
    return df

df = create_features(df)
```

```python
print(df.columns)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.boxplot(data=df, x='hour', y='AEP_MW',  palette="Set3")
# sns.boxplot(data=df, x='hour', y='AEP_MW', hue='hour', dodge=False, palette="tab20")
ax.set_title('MW by Hour')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.boxplot(data=df, x='month', y='AEP_MW', palette='Blues')
ax.set_title('MW by Month')
plt.show()

```

```python
train = create_features(train)
test = create_features(test)

FEATURES = ['dayofyear', 'hour', 'dayofweek', 'quarter', 'month', 'year']
TARGET = 'AEP_MW'

X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

```python
reg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',    
                       n_estimators=1000,
                       early_stopping_rounds=50,
                       objective='reg:linear',
                       max_depth=3,
                       learning_rate=0.01)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        verbose=100)

```

```python
fi = pd.DataFrame(data=reg.feature_importances_,
             index=reg.feature_names_in_,
             columns=['importance'])
fi.sort_values('importance').plot(kind='barh', title='Feature Importance')
plt.show()
```

```python
test['prediction'] = reg.predict(X_test)
# df.loc[X_test.index, 'prediction'] = reg.predict(X_test)
df = df.merge(test[['prediction']], how='left', left_index=True, right_index=True)


print(df.columns)
print(df.tail())
```

```python
ax = df[['AEP_MW']].plot(figsize=(15, 5))
df['prediction'].plot(ax=ax, style='.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Data and Prediction')
plt.show()
```

```python
ax = df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['AEP_MW'] \
    .plot(figsize=(15, 5), title='Week Of Data')
df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['prediction'] \
    .plot(style='.')
plt.legend(['Truth Data','Prediction'])
plt.show()
```

```python
score = np.sqrt(mean_squared_error(test['AEP_MW'], test['prediction']))
print(f'RMSE Score on Test set: {score:0.2f}')
```

```python
test['error'] = np.abs(test[TARGET] - test['prediction'])
test['date'] = test.index.date
test.groupby(['date'])['error'].mean().sort_values(ascending=False).head(10)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.data.preprocess import load_pjm_dataset

# --- Load target series (PJME) ---
# Load target series (PJME)
filename = "/Users/vkbharadwaj/Dropbox/Polaris/scratch/projects/AI_projects/from-skyrmions-to-smartgrids/data/PJM_dataset_Kaggle/pjm_hourly_est.csv"
df_pjme = load_pjm_dataset(filename, target_col="PJME")


# --- 1. Handle duplicate timestamps ---
duplicates = df_pjme.index.duplicated().sum()
print(f"Duplicate timestamps: {duplicates}")

# Option: average duplicates
df_pjme = df_pjme.groupby(df_pjme.index).mean()

# --- 2. Reindex to continuous hourly range ---
full_index = pd.date_range(start=df_pjme.index.min(), end=df_pjme.index.max(), freq="H")
df_pjme = df_pjme.reindex(full_index)

print("After reindexing:")
print("Missing values:", df_pjme.isna().sum())

# --- 3. Handle missing values (interpolation recommended) ---
df_interp = df_pjme.interpolate(method="time")

# --- 4. Detect outliers (z-score > 3) ---
z_scores = (df_interp - df_interp.mean()) / df_interp.std()
outliers = np.abs(z_scores) > 3
print("Outliers detected:", outliers.sum().values[0])

# Clip outliers to 1st99th percentile
df_cleaned = df_interp.clip(lower=df_interp.quantile(0.01)[0],
                            upper=df_interp.quantile(0.99)[0])

# --- 5. Save cleaned dataset ---
# df_cleaned.to_csv("data/processed/pjme_cleaned.csv")
print("Cleaned dataset saved to data/processed/pjme_cleaned.csv")

# --- 6. Quick plot ---
plt.figure(figsize=(12,4))
df_cleaned.iloc[:500].plot(legend=False, title="PJME Cleaned (first 500 hours)")
plt.show()
```

#### Step 3:  Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
filename = "../..//data/PJM_dataset_Kaggle/processed/pjme_cleaned.csv"
df = pd.read_csv(filename, parse_dates=["Unnamed: 0"], index_col="Unnamed: 0")
df.index.name = "Datetime"
print(df.info())
```

##### Step 3.1:  Basic Overview

```python
# Summary statistics

print(df.describe())

df.index = pd.to_datetime(df.index)
df_resampled = df.resample("D").mean()    # daily mean reduces 180k  ~7300 points

# Plot entire series
plt.figure(figsize=(15,4))
plt.plot(df_resampled, color="steelblue")
plt.title("PJME Hourly Energy Consumption (19982018)")
plt.ylabel("Load (MW)")
plt.xlabel("Time")
plt.show()


```

##### Step 3.2:  Rolling Statistics (Trend & Smoothness)

```python
window = 24*7  # 7-day rolling window
df["rolling_mean"] = df.iloc[:,0].rolling(window).mean()


df.index = pd.to_datetime(df.index)
df_resampled = df.resample("D").mean()    # daily mean reduces 180k  ~7300 points


plt.figure(figsize=(15,4))
plt.plot(df_resampled.iloc[:,0], alpha=0.4, label="Hourly Load")
plt.plot(df_resampled["rolling_mean"], color="red", label="7-Day Rolling Mean")
plt.legend()
plt.title("Rolling Average Trend")
plt.show()
```

##### Step 3.3:  Seasonality: Daily & Weekly Patterns

```python
df["hour"] = df.index.hour
df["dayofweek"] = df.index.dayofweek

# Average load by hour
hourly_pattern = df.groupby("hour")[df.columns[0]].mean()

plt.figure(figsize=(10,4))
hourly_pattern.plot(marker="o")
plt.title("Average Hourly Demand Pattern")
plt.xlabel("Hour of Day")
plt.ylabel("Load (MW)")
plt.show()

# Average load by weekday
weekday_pattern = df.groupby("dayofweek")[df.columns[0]].mean()
sns.barplot(x=weekday_pattern.index, y=weekday_pattern.values, palette="Blues_d")
plt.title("Average Load by Day of Week (0=Mon)")
plt.ylabel("Load (MW)")
plt.show()
```

##### Step 3.4:  Monthly / Yearly Trends

```python
df["month"] = df.index.month
df["year"] = df.index.year

# Monthly average
monthly_avg = df.groupby("month")[df.columns[0]].mean()
monthly_avg.plot(kind="bar", figsize=(10,4), color="teal")
plt.title("Average Monthly Energy Demand")
plt.xlabel("Month")
plt.ylabel("Load (MW)")
plt.show()

# Yearly trend
yearly_avg = df.groupby("year")[df.columns[0]].mean()
plt.figure(figsize=(12,4))
plt.plot(yearly_avg.index, yearly_avg.values, marker="o", color="darkorange")
plt.title("Yearly Average Load Trend")
plt.xlabel("Year")
plt.ylabel("Load (MW)")
plt.show()
```

##### Step 3.5:  Distribution & Outlier Check

```python
sns.histplot(df.iloc[:,0], bins=50, kde=True, color="royalblue")
plt.title("Distribution of Hourly Load Values")
plt.xlabel("Load (MW)")
plt.show()

sns.boxplot(x=df.iloc[:,0], color="skyblue")
plt.title("Boxplot of Load Values")
plt.show()
```

##### Step 3.6:  Summarise Plots

Perfect  Youve actually done **excellent work**  you now have *every key exploratory plot* that energy analysts and data scientists use to understand load patterns. Lets go through each one carefully so you fully understand what youre seeing and why it matters.

---

##### 1 **PJME Hourly Energy Consumption (19982018)**

 **What it shows:**

* The **entire time series** of hourly load (electricity demand) over ~20 years.
* You can see **clear repeating yearly patterns**  demand spikes in winters and summers (due to heating/cooling).

 **Why it matters:**

* Shows that the data is **non-stationary**  it has trends and seasonal cycles.
* This is why models like **ARIMA** need differencing, and **LSTMs/Transformers** must learn periodic dependencies.

---

##### 2 **Rolling Average Trend (7-Day Rolling Mean)**

 **What it shows:**

* The blue line = actual hourly demand.
* The red line = 7-day moving average (smooths out short fluctuations).

 **Why it matters:**

* You see a **repeating weekly cycle**  higher weekdays, lower weekends.
* Helps you confirm the natural periodicity of the data (~7 days).
* The smoothed red line is what youd expect a forecasting model to approximate.

---

##### 3 **Average Hourly Demand Pattern**

 **What it shows:**

* Average electricity consumption for each hour of the day (aggregated over all days).
* Load is lowest around **35 a.m.**, and peaks around **69 p.m.**

 **Why it matters:**

* Captures **daily seasonality** (the 24-hour pattern).
* Youll use this insight to decide your **input window size** for models (e.g., 24h = 1 full cycle).
* It reflects real human behavior  low night activity, high evening demand.

---

##### 4 **Average Load by Day of Week (0 = Monday)**

 **What it shows:**

* Average consumption per weekday.
* Weekdays (MonFri) have similar higher loads.
* Weekends (SatSun) show clear drop  less industrial/commercial demand.

 **Why it matters:**

* Confirms **weekly seasonality (7-day cycle)**.
* Suggests adding a **day-of-week feature** in your models to improve predictions.

---

##### 5 **Average Monthly Energy Demand**

 **What it shows:**

* Average load per month (112).
* Peaks around **JulyAugust** (summer AC demand) and **JanuaryFebruary** (heating).
* Dips around **AprilMay** and **OctoberNovember** (mild weather).

 **Why it matters:**

* Confirms **annual seasonality**.
* Useful for long-horizon forecasting and decomposition (trend + season + residual).

---

##### 6 **Yearly Average Load Trend**

 **What it shows:**

* Average load per year.
* Steady growth until ~2007, then slight decline/stabilization (possibly due to efficiency improvements, recession, or renewable adoption).

 **Why it matters:**

* Indicates **non-stationary trend** across years.
* Long-term drift is something neural networks must learn implicitly (or you can detrend data first).

---

##### 7 **Distribution of Hourly Load Values**

 **What it shows:**

* Histogram + KDE (density curve) of load values.
* Most loads fall between **25,00040,000 MW**, with a right-skewed tail.

 **Why it matters:**

* Confirms that load values are **not normally distributed** (slightly skewed).
* Suggests scaling (e.g., MinMaxScaler) is better than normalization for neural networks.

---

##### 8 **Boxplot of Load Values**

 **What it shows:**

* Median ~32,000 MW, with whiskers up to ~50,000 MW.
* The points beyond whiskers are **outliers** (rare high-demand peaks).

 **Why it matters:**

* Validates the outlier clipping you did earlier (1st99th percentile).
* These peaks correspond to extreme weather or major events  real but rare.

---

##### 9 **Autocorrelation Plot**

 **What it shows:**

* Correlation of load with itself at different time lags.
* The strong repeating oscillations show daily/weekly periodicity  peaks every 24 and 168 lags.

 **Why it matters:**

* Confirms strong autocorrelation  perfect for sequence models like **LSTM/GRU/Transformer**.
* Helps you choose `input_window`:

  * **24 lags** = 1-day pattern
  * **168 lags** = 1-week pattern

---

#####  Overall Takeaways from EDA

| Insight                 | Evidence             | Modeling Implication                                 |
| ----------------------- | -------------------- | ---------------------------------------------------- |
| **Daily seasonality**   | Hourly pattern plot  | Use 24-hour input windows                            |
| **Weekly seasonality**  | Day-of-week bar plot | Include weekday feature                              |
| **Annual seasonality**  | Monthly pattern plot | Include month or sine/cosine encoding                |
| **Non-stationarity**    | Yearly trend plot    | Detrending or differencing for ARIMA; scaling for DL |
| **Autocorrelation**     | ACF plot             | Confirms sequence learning is appropriate            |
| **Outliers exist**      | Boxplot              | Clip or robust-scale                                 |
| **Skewed distribution** | Histogram            | Use MinMaxScaler instead of StandardScaler           |

---

* The data is **periodic, non-stationary, and correlated**  perfect for **time-series deep learning**.
